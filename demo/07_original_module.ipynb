# 使用例２
System roleを設定する

```python
# System roleを設定しなかった場合
gpt = chatutil.NewChat()
reply = gpt.generate('バランスのいい朝食を考えて', temperature=0.0)
print(reply)
```

```python
# System roleを設定した場合
role = 'あなたはフレンチシェフです'
gpt.reset_role(role=role)
reply = gpt.generate('バランスのいい朝食を考えて', temperature=0.0)
print(reply)
```

```python
# System roleを再設定する
role = 'あなたは中華料理屋です'
gpt = chatutil.NewChat(role=role)
reply = gpt.generate('バランスのいい朝食を考えて', temperature=0.0)
print(reply)
```

```python
# System roleを消す
gpt.reset_role()
reply = gpt.generate('バランスのいい朝食を考えて', temperature=0.0)
print(reply)
```

# 使用例３
イテレーションさせる

```python
# インスタンスを生成
role = '私が入力する文字列にアルファベットが含まれていたら、それを小文字にして返して'
gpt = chatutil.NewChat(role=role)

before = ['AAA', 'BbB', 'ccc', 'D11']
after = []
for strings in before:
    # クレジットカードをOpenAIに登録していない場合は1分間のリクエスト回数が20回を超えないようにする
    # time.sleep(3)
    reply = gpt.generate(strings, temperature=0.0)
    after.append(reply)
    gpt.clear()
print(after)
```

# 使用例４
２つのChatGPTにレスバトルさせる

```python
# インスタンスを生成
role1 = 'あなたはGX(グリーントランスフォーメーション)の専門家です。GXについて私と議論し、より良い経営戦略を引き出してください。'
gpt1 = chatutil.NewChat(role=role1)
role2 = 'あなたは経営者です。GX(グリーントランスフォーメーション)について私と議論し、より良い経営戦略を考えてください。'
gpt2 = chatutil.NewChat(role=role2)

# 前準備
reply_list = ['まずはGXについて教えてください。']

for i in range(5):
    # 先攻
    msg = reply_list[-1]
    reply = gpt1.generate(msg)
    print(f'専門家:\n{reply}\n')
    reply_list.append(reply)
    # 後攻
    msg = reply_list[-1]
    reply = gpt2.generate(msg)
    print(f'経営者:\n{reply}\n')
    reply_list.append(reply)
```

```python
# やり取りを保存する
messages = gpt2.messages
content = [msg['content'] for msg in messages]
df = pd.DataFrame(content, columns=['chat'])
df.to_csv('./data/chat_history.csv', index=False, encoding='utf-8-sig')
```
